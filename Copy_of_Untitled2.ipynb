```python
import pandas as pd
df = pd.read_csv("student-por.csv")
```

```python
df.head()
```

```python
avg_score = df['G3'].mean()
avg_score

above_avg_students = df[df['G3']>avg_score]
above_avg_students[['G3']]

above = (df['G3'] > avg_score).sum()
print('Above average:',above)

bellow = (df['G3'] < avg_score).sum()
print('Bellow average:',bellow)

```

```python
# find the top 5 students

df.sort_values(by='G3', ascending=False).head()[['G1','G2','G3']]
```

```python
# compare simple test mark with average mark

df[['G1','G2','G3']].mean()
```

```python
# find how many students passed the final exam

passed = (df['G3'] >= 10).sum()
print("passed:",passed)
```

```python
# find he students with 0 absensess

df[df['absences']==0]
```

```python
# find the students who improved from G1 to G3

df[df['G3']>df['G1']][['G1','G3']]
```

```python
# find the students who performed consistently among G1,G2,G3

df[(df['G1']==df['G2']) & (df['G2'] == df['G3'])][['G1','G2','G3']]
```

```python
#matplotlib
#matplotlib tool is useful for visualization

import matplotlib.pyplot as plt

plt.figure()
plt.hist(df['G3'] , bins = 10)
plt.xlabel("G3")
plt.ylabel("Frequency")
plt.title("Histogram of G3")
plt.show()

```

```python
avg_score__by_gender = df.groupby("sex")['G3'].mean()
avg_score__by_gender

plt.figure()
avg_score__by_gender.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average final marks")
plt.title("Average final marksby gender")
plt.show()
```

```python
# sactter plot

plt.figure()
plt.scatter(df['studytime'],df['G3'])
plt.xlabel("studytime")
plt.ylabel("G3")
plt.title("scatter plot of G1 VS G3")
plt.show()
```

```python
# sactter plot

plt.figure()
plt.scatter(df['studytime'],df['G1'])
plt.xlabel("studytime")
plt.ylabel("G1")
plt.title("scatter plot of G1 VS G3")
plt.show()
```

```python
# boxplot
plt.figure()
plt.boxplot(df['G3'])
plt.xlabel("")
plt.ylabel("G3")
plt.title("spread of final scores")
plt.show()
```

```python
# line plot

avg_marks = df[['G1','G2','G3']].mean()

plt.figure()
plt.plot(avg_marks.index, avg_marks.values, marker = 'o')
plt.xlabel("subject")
plt.ylabel("average marks")
plt.title("Average marks of subject")
plt.show()
```

```python
# with bar chart we will find number of students who passed and failed
# who failed and passed final exam

pass_ = df['G3'] >= 10

plt.figure()
plt.bar(['pass','fail'],[pass_.sum(),(~pass_).sum()])

plt.xlabel("result")
plt.ylabel("no of students")
plt.title("number of students who passed and failed")
plt.show()
```

```python
# histogram _ absence distribution

import matplotlib.pyplot as plt

plt.figure()
plt.hist(df['absences'] , bins = 10, color = 'skyblue', edgecolor = 'black')
plt.xlabel("Absences")
plt.ylabel("Frequency")
plt.title("Histogram of Absences")
plt.show()
```
